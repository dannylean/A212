```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path
import a212data

%matplotlib inline
datadir = a212data.__path__[0]
stinkpath = Path(datadir).joinpath('stinkbug.png')
img=mpimg.imread(str(stinkpath))
imgplot = plt.imshow(img)
lum_img = img[:,:,0]
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('hot')
plt.colorbar()
plt.show()

cmap1=sns.cubehelix_palette(dark=0,light=1,reverse=True,as_cmap=True)
imgplot.set_cmap(cmap1)
plt.show()

cmap2=sns.light_palette("seagreen",reverse=True,as_cmap=True)
imgplot.set_cmap(cmap2)
plt.show()

cmap3=sns.dark_palette("seagreen",reverse=True,as_cmap=True)
imgplot.set_cmap(cmap3)
plt.show()
```
